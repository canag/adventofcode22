```python
items = {}
counter = {}
operations = {}
value_tests = {}
dest_true = {}
dest_false = {}
N = int(len(lines)/7)
for k in range(N):
    items[k] = []
    counter[k] = 0

# round 1
i = 0
while i<len(lines):
    k = int(lines[i][7:-1])
    #print(f'Monkey {k}:')
    items[k] = [int(s) for s in lines[i+1][18:].split(",")] + items[k]

    operations[k] = lines[i+2][13:]
    value_tests[k] = int(lines[i+3][21:])
    dest_true[k] = int(lines[i+4][29:])
    dest_false[k] = int(lines[i+5][30:])

    for item in items[k]:
        counter[k] += 1
        old = item
        exec(operations[k])
        #val = new // 3
        val = new 
        if val%value_tests[k]==0:
            dest = dest_true[k]
        else:
            dest = dest_false[k]
        items[dest].append(val)
    items[k] = []
    i += 7

print(items)
print(counter)

# round 2 to 20

for round in range(2, 10001):
    for k in range(N):
        for item in items[k]:
            counter[k] += 1
            old = item
            exec(operations[k])
            #val = new // 3
            val = new
            val = val % 9699690
            if val%value_tests[k]==0:
                dest = dest_true[k]
            else:
                dest = dest_false[k]
            items[dest].append(val)
        items[k] = []
    if round%1000==0:
        print(round, items)

print(round, counter)
print(np.prod(sorted(list(counter.values()))[-2:]))
```
