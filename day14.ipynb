```python
# day 14
f = open("data/data14.txt", "r")
lines = f.readlines()
lines = [l[:-1] for l in lines]
list_segments = []
for line in lines:
    seg = []
    for coord in line.split(" -> "):
        x, y = [int(v) for v in coord.split(",")]
        seg.append((x, y))
    list_segments.append(seg)
#print(list_segments)

list_rocks = []
for seg in list_segments:
    for i in range(len(seg)-1):
        ax, ay = seg[i]
        bx, by = seg[i+1]
        if ax==bx:
            tmp = [(ax, y) for y in range(min(ay, by), max(ay, by)+1)]
        elif ay==by:
            tmp = [(x, ay) for x in range(min(ax, bx), max(ax, bx)+1)]
        list_rocks.extend(tmp)        

part1=False
if part1:
    xmin, xmax = min([x[0] for x in list_rocks])-1, max([x[0] for x in list_rocks])+1
    ymin, ymax = min([x[1] for x in list_rocks]+[0]), max([x[1] for x in list_rocks])
else:
    ymin, ymax = min([x[1] for x in list_rocks]+[0]), max([x[1] for x in list_rocks])+2
    height = ymax-ymin
    xmin, xmax = min([x[0] for x in list_rocks])-3, max([x[0] for x in list_rocks])+3
    xmin, xmax = min(xmin, 500 - height), max(xmax, 500 + height)

    
A = [['.']*(xmax-xmin+1) for i in range(ymax-ymin+1)]
for coord in list_rocks:
    A[coord[1]-ymin][coord[0]-xmin] = '#'
for x in range(xmin, xmax+1):
    A[ymax-ymin][x-xmin] = '#'
list_x = list(range(xmin, xmax+1))
list_y = list(range(ymin, ymax+1))

print("  ", "".join([str(x)[0] for x in list_x]))
print("  ", "".join([str(x)[1] for x in list_x]))
print("  ", "".join([str(x)[2] for x in list_x]))

for i in range(len(A)):
    print(str(list_y[i]).zfill(2), "".join(A[i]))
```

```python
def get_next(pos):
    i = pos[1] - ymin
    j = pos[0] - xmin    
    if A[i+1][j]=='.':
        return pos[0], pos[1]+1
    elif A[i+1][j-1]=='.':
        return pos[0]-1, pos[1]+1
    elif A[i+1][j+1]=='.':
        return pos[0]+1, pos[1]+1
    else:
        return pos

falling = blocked = False
while (falling==False)&(blocked==False):
    pos = (500, 0)
    next_pos = get_next(pos)
    while pos != next_pos:
        pos = next_pos
        next_pos = get_next(pos)
        if next_pos[1] == ymax:
            falling=True
            break
    if falling == False:
        A[pos[1] - ymin][pos[0] - xmin] = 'o'
    if next_pos == (500, 0):
        blocked = True
        
print(" ", "".join([str(x)[0] for x in list_x]))
print(" ", "".join([str(x)[1] for x in list_x]))
print(" ", "".join([str(x)[2] for x in list_x]))

for i in range(len(A)):
    print(str(list_y[i]).zfill(2), "".join(A[i]))
len([s for a in A for s in a if s=='o'])
```
