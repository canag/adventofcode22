```python
import pandas as pd
import numpy as np

f = open("data3.txt", "r")
lines = f.readlines()
lines = [l[:-1] for l in lines]
df = pd.DataFrame(lines, columns=["input"])
df['len'] = df.input.str.len()
df['semilen'] = (df.len/2).astype('int')
df.head()
```

```python
df['part1'] = df.apply(lambda x: x['input'][:x['semilen']], axis=1)
df['part2'] = df.apply(lambda x: x['input'][x['semilen']:], axis=1)
df['common'] = df.apply(lambda x: list(set(x['part1'])&set(x['part2']))[0], axis=1)
df.head()
```

```python
ord('a'), ord('z'), ord('A'), ord('Z')
```

```python
def f(c):
    n = ord(c)
    if n>96: # lowercase
        n -= 96
    else: # uppercase
        n -= (64-26)
    return n
df['priority'] = df.apply(lambda x: f(x['common']), axis=1)
df.head()
```

```python
df['priority'].sum()
```

```python
df.drop(["len", "semilen", "part1", "part2", "common", "priority"], axis=1, inplace=True)
df.head()
```

```python
df['input2'] = df.input.shift(-1)
df['input3'] = df.input.shift(-2)
df = df.iloc[::3]
df.head()
```

```python
df['common'] = df.apply(lambda x: list(set(x['input'])&set(x['input2'])&set(x['input3']))[0], axis=1)

df.head()
```
