```python
import pandas as pd
import numpy as np

f = open("data4.txt", "r")
lines = f.readlines()
lines = [l[:-1] for l in lines]
df = pd.DataFrame(lines, columns=["input"])
df.head()
```

```python
df["set1_a"] = df.input.str.split(",").str[0].str.split("-").str[0].astype('int')
df["set1_b"] = df.input.str.split(",").str[0].str.split("-").str[1].astype('int')
df["set1"] = df.apply(lambda x: set(range(x['set1_a'], x['set1_b']+1)), axis=1)
df["set2_a"] = df.input.str.split(",").str[1].str.split("-").str[0].astype('int')
df["set2_b"] = df.input.str.split(",").str[1].str.split("-").str[1].astype('int')
df["set2"] = df.apply(lambda x: set(range(x['set2_a'], x['set2_b']+1)), axis=1)
df.drop(["set1_a", "set1_b", "set2_a", "set2_b"], axis=1, inplace=True)
df["set1inset2"] = df.apply(lambda x: x["set1"].issubset(x["set2"]), axis=1)
df["set2inset1"] = df.apply(lambda x: x["set2"].issubset(x["set1"]), axis=1)
df["cond"] = df.set1inset2 | df.set2inset1
print(df.cond.sum())
df.head()
```

```python
df["overlap"] = df.apply(lambda x: len(x["set1"]&x["set2"]), axis=1)
df.head()
```
